### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Create a data dictionary that looks like the DataFrame below

```python
data = {
'evolution': ['Bulbasaur', 'Charmander', 'Squirtle', 'Pidgey', 'Pikachu', 'Eevee', 'Clefairy', 'Jigglypuff', 'Gastly', 'Onix', 'Machop', 'Geodude', 'Ditto', 'Seel', 'Drowzee'],
    'hp': [45, 39, 44, 40, 35, 55, 70, 115, 30, 35, 70, 40, 48, 65, 60],
    'name': ['Ivysaur', 'Charmeleon', 'Wartortle', 'Pidgeotto', 'Raichu', 'Vaporeon', 'Clefable', 'Wigglytuff', 'Haunter', 'Steelix', 'Machoke', 'Graveler', 'Ditto', 'Dewgong', 'Hypno'],
    'pokedex': [2, 5, 8, 17, 25, 133, 35, 40, 92, 95, 67, 75, 132, 86, 96],
    'type': ['Grass/Poison', 'Fire', 'Water', 'Normal/Flying', 'Electric', 'Water', 'Fairy', 'Normal/Fairy', 'Ghost/Poison', 'Rock/Ground', 'Fighting', 'Rock/Ground', 'Normal', 'Water/Ice', 'Psychic']
}
```

### Step 3. Assign it to a variable called pokemon

```python
pokemon = pd.DataFrame(data)
pokemon
```

### Step 4. Ops...it seems the DataFrame columns are in alphabetical order. Place  the order of the columns as name, type, hp, evolution, pokedex

```python
new_order = ['name', 'type', 'hp', 'evolution', 'pokedex']
pokemon = pokemon[new_order]
pokemon
```

### Step 5. Add another column called place, and insert what you have in mind.

```python
# Create a new series for 'place'
place = pd.Series(['Forest', 'Mountain', 'Water', 'Forest', 'Forest', 'Water', 'Cave', 'Grassland', 'Haunted Mansion', 'Cave', 'Mountain', 'Mountain', 'Ditto Habitat', 'Sea', 'Dream World'], name='place')

# Make a copy of the DataFrame and use .loc on the copy
pokemon_copy = pokemon.copy()
pokemon_copy.loc[:, 'place'] = place
pokemon = pokemon_copy
pokemon
```

#### what is the most pokemon type

```python
# Use value_counts to count the occurrences of each type
type_counts = pokemon['type'].value_counts()
max_count = type_counts.max()
most_repeated_types = type_counts[type_counts == max_count]
most_repeated_type_list = most_repeated_types.index.tolist()
print(most_repeated_type_list)
```
